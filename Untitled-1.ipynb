```python
# Download dataset locally — run once, skips already downloaded files
LOCAL_DIR = Path('autonomic_aging_data')
LOCAL_DIR.mkdir(exist_ok=True)

BASE_URL   = f'https://physionet.org/files/{PHYSIONET_DB}/1.0.0'
extensions = ['.hea', '.dat']
downloaded = 0; skipped = 0; failed = []

for rid in tqdm(record_list, desc='Downloading records'):
    for ext in extensions:
        fname    = f'{rid}{ext}'
        out_path = LOCAL_DIR / fname
        if out_path.exists():
            skipped += 1; continue
        try:
            urllib.request.urlretrieve(f'{BASE_URL}/{fname}', out_path)
            downloaded += 1
            time.sleep(0.05)
        except Exception:
            failed.append(fname)

print(f'✅ Downloaded : {downloaded} files')
print(f'   Skipped    : {skipped} (already existed)')
print(f'   Failed     : {len(failed)}')
```

```python
def extract_features_local(record_id, analysis_sec=ANALYSIS_SEC):
    try:
        rec = wfdb.rdrecord(str(LOCAL_DIR / str(record_id)))
        n   = min(analysis_sec * FS, rec.p_signal.shape[0])

        sig_lower = [s.lower() for s in rec.sig_name]
        ecg_idx   = next((i for i, s in enumerate(sig_lower) if 'ecg' in s), 0)
        bp_idx    = next((i for i, s in enumerate(sig_lower)
                          if any(k in s for k in ['bp','abp','pressure','nibp'])), 2)

        ecg = rec.p_signal[:n, ecg_idx]
        bp  = rec.p_signal[:n, bp_idx]

        # ── ECG: clean + R-peak detection ─────────────────────
        ecg_c = nk.ecg_clean(ecg, sampling_rate=FS)
        _, rp = nk.ecg_peaks(ecg_c, sampling_rate=FS)
        r_idx = rp['ECG_R_Peaks']

        if len(r_idx) < 30:
            return None

        rr_ms = np.diff(r_idx) / FS * 1000

        # ── Time-domain ────────────────────────────────────────
        td   = nk.hrv_time(rp, sampling_rate=FS, show=False)
        sdnn  = np.std(rr_ms, ddof=1)
        rmssd = np.sqrt(np.mean(np.diff(rr_ms)**2))
        mean_rr = np.mean(rr_ms)
        cv    = sdnn / mean_rr * 100   # coefficient of variation

        pnn50 = td.get('HRV_pNN50', [np.nan])[0]
        sdsd  = td.get('HRV_SDSD',  [np.nan])[0]

        # ── Frequency-domain ───────────────────────────────────
        fd = nk.hrv_frequency(rp, sampling_rate=FS, show=False)
        lf_power    = fd.get('HRV_LF',   [np.nan])[0]
        hf_power    = fd.get('HRV_HF',   [np.nan])[0]
        lf_hf       = fd.get('HRV_LFHF', [np.nan])[0]
        total_power = fd.get('HRV_TP',   [np.nan])[0]
        vlf_power   = fd.get('HRV_VLF',  [np.nan])[0]

        # ── Nonlinear ──────────────────────────────────────────
        nl = nk.hrv_nonlinear(rp, sampling_rate=FS, show=False)
        cd  = nl.get('HRV_CD',        [np.nan])[0]
        if np.isnan(cd):
            cd = nl.get('HRV_SampEn', [np.nan])[0]
        dfa_a1 = nl.get('HRV_DFA_alpha1', [np.nan])[0]
        dfa_a2 = nl.get('HRV_DFA_alpha2', [np.nan])[0]
        sd1    = nl.get('HRV_SD1',        [np.nan])[0]
        sd2    = nl.get('HRV_SD2',        [np.nan])[0]
        samp_en = nl.get('HRV_SampEn',    [np.nan])[0]

        # ── BP-derived ─────────────────────────────────────────
        bp_clean = bp[~np.isnan(bp)]
        bp_mean  = np.mean(bp_clean)
        bp_std   = np.std(bp_clean)
        sbp_est  = np.percentile(bp_clean, 95)
        dbp_est  = np.percentile(bp_clean, 5)
        pp       = sbp_est - dbp_est

        return {
            'record_id':   int(record_id),
            # Phase 1 baseline (paper replication)
            'sdnn':        sdnn,
            'rmssd':       rmssd,
            'cd':          cd,
            # Phase 2 extended HRV
            'mean_rr':     mean_rr,
            'cv':          cv,
            'pnn50':       pnn50,
            'sdsd':        sdsd,
            'lf_power':    lf_power,
            'hf_power':    hf_power,
            'lf_hf':       lf_hf,
            'total_power': total_power,
            'vlf_power':   vlf_power,
            'dfa_alpha1':  dfa_a1,
            'dfa_alpha2':  dfa_a2,
            'sd1':         sd1,
            'sd2':         sd2,
            'samp_en':     samp_en,
            # Phase 2 BP features
            'bp_mean':     bp_mean,
            'bp_std':      bp_std,
            'sbp_est':     sbp_est,
            'dbp_est':     dbp_est,
            'pulse_pressure': pp,
        }

    except Exception:
        return None

# Test
test = extract_features_local(SAMPLE_ID)
print('✅ Works!' if test else '❌ Failed')
if test:
    print(f'   Features extracted : {len(test) - 1}')
    print(f'   SDNN       : {test["sdnn"]:.2f} ms')
    print(f'   RMSSD      : {test["rmssd"]:.2f} ms')
    print(f'   CD         : {test["cd"]:.4f}')
    print(f'   LF/HF      : {test["lf_hf"]:.3f}')
    print(f'   BP mean    : {test["bp_mean"]:.1f} mmHg')
    print(f'   Pulse pres : {test["pulse_pressure"]:.1f} mmHg')
```

```python
FEATURE_CSV  = OUTPUT_DIR / 'features_all.csv'
all_features = []
failed       = []

for rid in tqdm(record_list, desc='Extracting features'):
    result = extract_features_local(rid)
    if result:
        all_features.append(result)
    else:
        failed.append(rid)

features_df = pd.DataFrame(all_features)
features_df.to_csv(FEATURE_CSV, index=False)

print(f'\n✅ Done!')
print(f'   Successful : {len(all_features)} / {len(record_list)}')
print(f'   Failed     : {len(failed)} records')
if failed:
    print(f'   Failed IDs : {failed[:10]}')
display(features_df.head())
display(features_df.describe().round(2))
```

```python
# ── 2.3  Merge with metadata ───────────────────────────────────
features_df['record_id'] = features_df['record_id'].astype(int)
metadata['id']           = metadata['id'].astype(int)

df = features_df.merge(
        metadata[['id', 'age_group', 'age_label', 'age_broad',
                  'age_mid', SEX_COL, BMI_COL]],
        left_on='record_id', right_on='id', how='inner')

print(f'After merge: {len(df)} subjects')

# ── 2.4  Remove physiologically impossible outliers ────────────
# HRV values outside these ranges indicate failed R-peak detection
HRV_LIMITS = {
    'sdnn':    (1,   300),    # ms
    'rmssd':   (1,   300),    # ms
    'mean_rr': (400, 1500),   # ms  → HR between 40–150 bpm
    'pnn50':   (0,   100),    # %
    'cd':      (0.1, 2.5),
}

before = len(df)
for col, (lo, hi) in HRV_LIMITS.items():
    df = df[(df[col] >= lo) & (df[col] <= hi)]

print(f'Removed {before - len(df)} outlier subjects')
print(f'Clean dataset: {len(df)} subjects')

# ── 2.5  Handle remaining missing values ───────────────────────
FEATURE_COLS = ['sdnn','rmssd','cd','mean_rr','cv','pnn50','sdsd',
                'lf_power','hf_power','lf_hf','total_power','vlf_power',
                'dfa_alpha1','dfa_alpha2','sd1','sd2','samp_en',
                'bp_mean','bp_std','sbp_est','dbp_est','pulse_pressure']

# Median imputation per age group for any remaining NaNs
for col in FEATURE_COLS:
    if df[col].isnull().any():
        df[col] = df.groupby('age_broad')[col].transform(
            lambda x: x.fillna(x.median()))

df = df.dropna(subset=FEATURE_COLS)
df.to_csv(OUTPUT_DIR / 'dataset.csv', index=False)

print(f'\n✅ Final clean dataset: {len(df)} subjects × {len(FEATURE_COLS)} features')
print(f'\nAge group distribution after cleaning:')
display(df['age_broad'].value_counts().reindex(age_order).to_frame())
display(df[FEATURE_COLS].describe().round(2))
```

```python
# ── PCA ────────────────────────────────────────────────────────
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

FEATURE_COLS = ['sdnn','rmssd','cd','mean_rr','cv','pnn50','sdsd',
                'lf_power','hf_power','lf_hf','total_power','vlf_power',
                'dfa_alpha1','dfa_alpha2','sd1','sd2','samp_en',
                'bp_mean','bp_std','sbp_est','dbp_est','pulse_pressure']

X      = df[FEATURE_COLS].values
y      = df['age_group'].values        # ordinal 1–15
groups = df['age_broad'].values

scaler = StandardScaler()
X_s    = scaler.fit_transform(X)

# Fit PCA
pca       = PCA()
X_pca     = pca.fit_transform(X_s)
explained = pca.explained_variance_ratio_
cumulative = np.cumsum(explained)

# How many components explain 90% variance?
n_90 = np.argmax(cumulative >= 0.90) + 1
print(f'Components to explain 90% variance: {n_90}')

# Scree plot
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].bar(range(1, len(explained)+1), explained*100,
            color='#2196F3', edgecolor='white')
axes[0].plot(range(1, len(explained)+1), cumulative*100,
             'r-o', markersize=4, label='Cumulative')
axes[0].axhline(90, color='gray', ls='--', lw=1, label='90%')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Explained Variance (%)')
axes[0].set_title('PCA Scree Plot', fontweight='bold')
axes[0].legend()

# PCA biplot — PC1 vs PC2 coloured by age group
for g in age_order:
    mask = groups == g
    axes[1].scatter(X_pca[mask, 0], X_pca[mask, 1],
                    alpha=0.4, s=15, color=AGE_COLORS[g], label=g)
axes[1].set_xlabel(f'PC1 ({explained[0]*100:.1f}%)')
axes[1].set_ylabel(f'PC2 ({explained[1]*100:.1f}%)')
axes[1].set_title('PCA — PC1 vs PC2 by Age Group', fontweight='bold')
axes[1].legend(fontsize=8)

plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'step3_pca.png', bbox_inches='tight')
plt.show()
```

```python
from sklearn.linear_model    import RidgeClassifier
from sklearn.ensemble        import RandomForestClassifier
from sklearn.neural_network  import MLPClassifier
from sklearn.svm             import SVC
from sklearn.preprocessing   import StandardScaler, LabelEncoder
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics         import (classification_report, confusion_matrix,
                                      f1_score, ConfusionMatrixDisplay)
from sklearn.pipeline        import Pipeline as SkPipeline
import xgboost as xgb

df = pd.read_csv(OUTPUT_DIR / 'dataset_model.csv')

FEATURE_COLS = ['cd','pnn50','samp_en','dfa_alpha2','cv','sd2','sdnn',
                'lf_hf','rmssd','sd1','sdsd','vlf_power','bp_std']

X  = df[FEATURE_COLS].values

# ✅ One encoder, fit on unique classes only
le     = LabelEncoder().fit(df['age_broad'])
y      = le.transform(df['age_broad'])
labels = le.classes_   # ✅ this is ['<30','30-39','40-49','50-59','60-69','≥70'] — 6 items only

scaler = StandardScaler()
X_s    = scaler.fit_transform(X)

print(f'✅ Ready: {X.shape[0]} subjects, {X.shape[1]} features, {len(labels)} classes')
print(f'Class labels : {list(labels)}')
print(f'Unique y values : {sorted(set(y))}')   # sanity check — should be [0,1,2,3,4,5]
print(f'\nClass distribution:')
display(pd.Series(le.inverse_transform(y)).value_counts()
          .reindex(age_order).to_frame('count'))
```

```python
# ── 4.2  5-Fold CV — all models inside proper Pipeline ────────


cv     = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
models = {
    'Ridge Classifier': SkPipeline([
        ('scaler', StandardScaler()),
        ('model',  RidgeClassifier(class_weight='balanced'))
    ]),
    'Random Forest': SkPipeline([
        ('scaler', StandardScaler()),
        ('model',  RandomForestClassifier(n_estimators=200,
                       class_weight='balanced',
                       random_state=42, n_jobs=-1))
    ]),
    'XGBoost': SkPipeline([
        ('scaler', StandardScaler()),
        ('model',  xgb.XGBClassifier(n_estimators=300,
                       learning_rate=0.05, max_depth=6,
                       eval_metric='mlogloss',
                       random_state=42))
    ]),
    'SVM': SkPipeline([
        ('scaler', StandardScaler()),
        ('model',  SVC(kernel='rbf', class_weight='balanced',
                       random_state=42))
    ]),
    'MLP': SkPipeline([
        ('scaler', StandardScaler()),
        ('model',  MLPClassifier(hidden_layer_sizes=(128, 64, 32),
                       activation='relu', max_iter=500,
                       early_stopping=True, random_state=42))
    ]),
}

cv_results = {}
for name, pipeline in models.items():
    scores = cross_validate(pipeline, X, y, cv=cv,
                            scoring=['f1_macro', 'accuracy'],
                            return_train_score=False)
    cv_results[name] = {
        'Macro F1 (mean)': scores['test_f1_macro'].mean().round(3),
        'Macro F1 (std)':  scores['test_f1_macro'].std().round(3),
        'Accuracy (mean)': scores['test_accuracy'].mean().round(3),
        'Accuracy (std)':  scores['test_accuracy'].std().round(3),
    }
    print(f'✅ {name} — F1: {scores["test_f1_macro"].mean():.3f} '
          f'± {scores["test_f1_macro"].std():.3f}')

results_df = pd.DataFrame(cv_results).T.sort_values('Macro F1 (mean)', ascending=False)
print('\n📊 5-Fold CV Results (6-class, no SMOTE):')
display(results_df)
results_df.to_csv(OUTPUT_DIR / 'model_comparison.csv')

# ✅ Named variable — not overwritten later
best_name_6class  = results_df.index[0]
best_model_6class = models[best_name_6class]
print(f'\nBest (6-class, no SMOTE) : {best_name_6class}')
```

```python
# ── 4.7  Retrain all 5 models on SMOTE-balanced data ──────────
# Important: SMOTE must be applied INSIDE cross-validation
# to avoid data leakage — use imblearn Pipeline
from imblearn.pipeline       import Pipeline as ImbPipeline
from imblearn.over_sampling  import SMOTE
from sklearn.model_selection import cross_validate, StratifiedKFold

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

models_smote = {
    'Ridge Classifier': RidgeClassifier(),
    'Random Forest':    RandomForestClassifier(n_estimators=200,
                            random_state=42, n_jobs=-1),
    'XGBoost':          xgb.XGBClassifier(n_estimators=300,
                            learning_rate=0.05, max_depth=6,
                            eval_metric='mlogloss', random_state=42),
    'SVM':              SVC(kernel='rbf', random_state=42),
    'MLP':              MLPClassifier(hidden_layer_sizes=(128, 64, 32),
                            activation='relu', max_iter=500,
                            early_stopping=True, random_state=42),
}

pipelines_smote = {
    name: ImbPipeline([
        ('smote',  SMOTE(random_state=42, k_neighbors=5)),
        ('scaler', StandardScaler()),
        ('model',  model)
    ])
    for name, model in models_smote.items()
}

cv_results_smote = {}
for name, pipe in pipelines_smote.items():
    scores = cross_validate(pipe, X, y, cv=cv,
                            scoring=['f1_macro', 'accuracy'],
                            return_train_score=False)
    cv_results_smote[name] = {
        'Macro F1 (mean)': round(scores['test_f1_macro'].mean(), 3),
        'Macro F1 (std)':  round(scores['test_f1_macro'].std(),  3),
        'Accuracy (mean)': round(scores['test_accuracy'].mean(), 3),
        'Accuracy (std)':  round(scores['test_accuracy'].std(),  3),
    }
    print(f'✅ {name} — F1: {scores["test_f1_macro"].mean():.3f} '
          f'± {scores["test_f1_macro"].std():.3f}')

results_smote_df = pd.DataFrame(cv_results_smote).T\
                     .sort_values('Macro F1 (mean)', ascending=False)
print(f'\n📊 Results WITH SMOTE (6-class):')
display(results_smote_df)
results_smote_df.to_csv(OUTPUT_DIR / 'model_comparison_smote.csv')

# ✅ Named variable — separate from 6-class no-SMOTE result
best_name_6class_smote = results_smote_df.index[0]
print(f'\nBest (6-class, SMOTE) : {best_name_6class_smote}')
```

```python
# ── Retrain all 5 models with 3 classes + SMOTE ────────────────
from imblearn.pipeline       import Pipeline as ImbPipeline
from imblearn.over_sampling  import SMOTE
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.preprocessing   import LabelEncoder

FEATURE_COLS = ['cd','pnn50','samp_en','dfa_alpha2','cv','sd2','sdnn',
                'lf_hf','rmssd','sd1','sdsd','vlf_power','bp_std']

X3      = df[FEATURE_COLS].values
le3     = LabelEncoder()
y3      = le3.fit_transform(df['age_3class'])
labels3 = le3.classes_
cv5     = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print(f'Classes: {list(labels3)}')
print(f'Distribution: {dict(zip(labels3, np.bincount(y3)))}')

models3 = {
    'Ridge Classifier': RidgeClassifier(),
    'Random Forest':    RandomForestClassifier(n_estimators=200,
                            random_state=42, n_jobs=-1),
    'XGBoost':          xgb.XGBClassifier(n_estimators=300,
                            learning_rate=0.05, max_depth=6,
                            eval_metric='mlogloss', random_state=42),
    'SVM':              SVC(kernel='rbf', random_state=42),
    'MLP':              MLPClassifier(hidden_layer_sizes=(128, 64, 32),
                            activation='relu', max_iter=500,
                            early_stopping=True, random_state=42),
}

pipelines3 = {
    name: ImbPipeline([
        ('smote',  SMOTE(random_state=42, k_neighbors=5)),
        ('scaler', StandardScaler()),
        ('model',  model)
    ])
    for name, model in models3.items()
}

cv_results3 = {}
for name, pipe in pipelines3.items():
    scores = cross_validate(pipe, X3, y3, cv=cv5,
                            scoring=['f1_macro', 'accuracy'],
                            return_train_score=False)
    cv_results3[name] = {
        'Macro F1 (mean)': round(scores['test_f1_macro'].mean(), 3),
        'Macro F1 (std)':  round(scores['test_f1_macro'].std(),  3),
        'Accuracy (mean)': round(scores['test_accuracy'].mean(), 3),
        'Accuracy (std)':  round(scores['test_accuracy'].std(),  3),
    }
    print(f'✅ {name} — F1: {scores["test_f1_macro"].mean():.3f} '
          f'± {scores["test_f1_macro"].std():.3f}')

results3_df = pd.DataFrame(cv_results3).T\
                .sort_values('Macro F1 (mean)', ascending=False)
print(f'\n📊 3-Class Results WITH SMOTE:')
display(results3_df)
results3_df.to_csv(OUTPUT_DIR / 'model_comparison_3class.csv')

# ✅ Named variable — separate from all previous experiments
best_name_3class = results3_df.index[0]
print(f'\nBest (3-class, SMOTE) : {best_name_3class}')
```

```python
# ── Create ECG × BP interaction features ──────────────────────
df_model = df.copy()

# 1. Baroreflex sensitivity proxy
#    BRS = HRV / BP variability — how well HR responds to BP changes
df_model['brs_proxy']     = df_model['rmssd'] / (df_model['bp_std'] + 1e-6)

# 2. Autonomic complexity index
#    CD × bp_std — signal complexity weighted by BP fluctuation
df_model['auto_complex']  = df_model['cd'] * df_model['bp_std']

# 3. Sympathovagal-BP coupling
#    LF/HF ratio weighted by BP variability
df_model['lfhf_bp']       = df_model['lf_hf'] * df_model['bp_std']

# 4. Cardiac efficiency index
#    pNN50 / bp_std — parasympathetic tone relative to BP instability
df_model['cardiac_eff']   = df_model['pnn50'] / (df_model['bp_std'] + 1e-6)

# 5. Complexity-autonomic index
#    samp_en × cv — entropy weighted by heart rate variability coefficient
df_model['complex_auto']  = df_model['samp_en'] * df_model['cv']

interaction_features = ['brs_proxy','auto_complex','lfhf_bp',
                        'cardiac_eff','complex_auto']

print('✅ Interaction features created')
print('\nCorrelation with age group code:')
corr = df_model[interaction_features + ['age_group']].corr()['age_group']\
         .drop('age_group').abs().sort_values(ascending=False)
display(corr.round(3).to_frame('|r| with age'))
```

```python
# ── Check if interaction features improve correlation ──────────
# Compare against original features
original_corr = df[['cd','pnn50','samp_en','dfa_alpha2','cv',
                     'bp_std','age_group']].corr()['age_group']\
                  .drop('age_group').abs().sort_values(ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].barh(original_corr.index, original_corr.values,
             color='#2196F3', edgecolor='white')
axes[0].set_title('Original Features\n|r| with Age', fontweight='bold')
axes[0].set_xlabel('|Pearson r|')
axes[0].axvline(0.1, color='gray', ls='--', lw=1)

int_corr = corr
axes[1].barh(int_corr.index, int_corr.values,
             color='#E91E63', edgecolor='white')
axes[1].set_title('Interaction Features\n|r| with Age', fontweight='bold')
axes[1].set_xlabel('|Pearson r|')
axes[1].axvline(0.1, color='gray', ls='--', lw=1)

fig.suptitle('Original vs Interaction Feature Correlation with Age',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'step4_interaction_correlation.png', bbox_inches='tight')
plt.show()
```

```python
# ── Retrain best model with interaction features added ─────────
FEATURE_COLS_INT = ['cd','pnn50','samp_en','dfa_alpha2','cv','sd2','sdnn',
                    'lf_hf','rmssd','sd1','sdsd','vlf_power','bp_std',
                    # interaction features
                    'brs_proxy','auto_complex','lfhf_bp',
                    'cardiac_eff','complex_auto']

X_int = df_model[FEATURE_COLS_INT].values
y3    = LabelEncoder().fit_transform(df_model['age_3class'])

pipe_int = ImbPipeline([
    ('smote',  SMOTE(random_state=42, k_neighbors=5)),
    ('scaler', StandardScaler()),
    ('model',  MLPClassifier(hidden_layer_sizes=(128,64,32),
                             activation='relu', max_iter=500,
                             early_stopping=True, random_state=42))
])

scores_int = cross_validate(pipe_int, X_int, y3, cv=cv5,
                             scoring=['f1_macro','accuracy'],
                             return_train_score=False)

print('── MLP with interaction features ──')
print(f'  Macro F1 : {scores_int["test_f1_macro"].mean():.3f} '
      f'± {scores_int["test_f1_macro"].std():.3f}')
print(f'  Accuracy : {scores_int["test_accuracy"].mean():.3f}')
print(f'\nPrevious MLP without interactions:')
print(f'  Macro F1 : {cv_results3["MLP"]["Macro F1 (mean)"]:.3f}')
print(f'\nImprovement: '
      f'{scores_int["test_f1_macro"].mean() - cv_results3["MLP"]["Macro F1 (mean)"]:.3f}')
```

```python
# ── 4.6b  Statistical significance of interaction features ────
from scipy import stats

# Get per-fold F1 scores for MLP without and with interaction features
scores_base = cross_validate(
    ImbPipeline([
        ('smote',  SMOTE(random_state=42, k_neighbors=5)),
        ('scaler', StandardScaler()),
        ('model',  MLPClassifier(hidden_layer_sizes=(128, 64, 32),
                                 activation='relu', max_iter=500,
                                 early_stopping=True, random_state=42))
    ]),
    X3, y3, cv=cv5, scoring='f1_macro'
)

scores_int_cv = cross_validate(
    ImbPipeline([
        ('smote',  SMOTE(random_state=42, k_neighbors=5)),
        ('scaler', StandardScaler()),
        ('model',  MLPClassifier(hidden_layer_sizes=(128, 64, 32),
                                 activation='relu', max_iter=500,
                                 early_stopping=True, random_state=42))
    ]),
    X_int, y3, cv=cv5, scoring='f1_macro'
)

f1_base = scores_base['test_score']
f1_int  = scores_int_cv['test_score']

# ✅ Paired t-test — same folds, so paired is correct
t_stat, p_value = stats.ttest_rel(f1_int, f1_base)

print('── Paired t-test: MLP with vs without interaction features ──')
print(f'   Fold F1 (baseline)    : {f1_base.round(3)}  mean={f1_base.mean():.3f}')
print(f'   Fold F1 (interaction) : {f1_int.round(3)}  mean={f1_int.mean():.3f}')
print(f'   t-statistic : {t_stat:.3f}')
print(f'   p-value     : {p_value:.4f}')

if p_value < 0.05:
    print('✅ Improvement is statistically significant (p < 0.05)')
else:
    print('⚠️  Improvement is NOT statistically significant (p ≥ 0.05)')
    print('   Interaction features may not be adding reliable value.')
```

```python
# --- Final SHAP analysis with exact feature names and class names ---

import shap
import pandas as pd
from sklearn.model_selection import train_test_split

# exact feature names from your notebook
FEATURE_COLS_INT = [
    'cd','pnn50','samp_en','dfa_alpha2','cv','sd2','sdnn',
    'lf_hf','rmssd','sd1','sdsd','vlf_power','bp_std',
    'brs_proxy','auto_complex','lfhf_bp',
    'cardiac_eff','complex_auto'
]

# rebuild dataframe with exact feature names
X_int_df = df_model[FEATURE_COLS_INT].copy()

# class names from your fitted label encoder
le3 = LabelEncoder().fit(df_model['age_3class'])
class_names = list(le3.classes_)

print("Class mapping:")
for i, name in enumerate(class_names):
    print(f"  Class {i}: {name}")

# split data
X_train_df, X_test_df, y_train, y_test = train_test_split(
    X_int_df, y3,
    test_size=0.2,
    stratify=y3,
    random_state=42
)

# make sure final best model is updated from final comparison
FINAL_BEST_NAME = final_df.index[0]
FINAL_BEST_PIPE = all_pipes_int[FINAL_BEST_NAME]

# fit final pipeline
FINAL_BEST_PIPE.fit(X_train_df.values, y_train)

# extract trained scaler and model
scaler = FINAL_BEST_PIPE.named_steps['scaler']
model = FINAL_BEST_PIPE.named_steps['model']

# transform using scaler
X_train_scaled = scaler.transform(X_train_df.values)
X_test_scaled = scaler.transform(X_test_df.values)

# restore dataframe with REAL feature names
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=FEATURE_COLS_INT)

print(f"\nFinal SHAP model: {type(model).__name__}")
print(f"Number of features: {len(FEATURE_COLS_INT)}")

# SHAP for XGBoost
if 'xgb' in type(model).__name__.lower():
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X_test_scaled)

    # bar plot
    shap.summary_plot(
        shap_values,
        X_test_scaled_df,
        plot_type='bar',
        class_names=class_names,
        show=True
    )


# SHAP for MLP fallback
elif 'mlp' in type(model).__name__.lower():
    background = shap.sample(X_train_scaled, min(100, len(X_train_scaled)), random_state=42)
    explainer = shap.KernelExplainer(model.predict_proba, background)

    X_sample = X_test_scaled[:min(100, len(X_test_scaled))]
    X_sample_df = pd.DataFrame(X_sample, columns=FEATURE_COLS_INT)

    shap_values = explainer.shap_values(X_sample)

    shap.summary_plot(
        shap_values,
        X_sample_df,
        plot_type='bar',
        class_names=class_names,
        show=True
    )


else:
    print("Current final model is neither XGBoost nor MLP. Add model-specific SHAP handling if needed.")
```

```python
# ── Step 7: Unsupervised Clustering ───────────────────────────
from sklearn.cluster     import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics     import silhouette_score, adjusted_rand_score
import umap

FEATURE_COLS = ['cd','pnn50','samp_en','dfa_alpha2','cv','sd2','sdnn',
                'lf_hf','rmssd','sd1','sdsd','vlf_power','bp_std',
                'brs_proxy','auto_complex','lfhf_bp',
                'cardiac_eff','complex_auto']

X_clust     = StandardScaler().fit_transform(df_model[FEATURE_COLS].values)
true_labels = LabelEncoder().fit_transform(df_model['age_3class'])
```

```python
# ── Elbow + Silhouette to choose k ────────────────────────────
inertias   = []
sil_scores = []
K_range    = range(2, 9)

for k in K_range:
    km     = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_clust)
    inertias.append(km.inertia_)
    sil_scores.append(silhouette_score(X_clust, labels))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(K_range, inertias, 'o-', color='#2196F3')
axes[0].set_title('Elbow Plot', fontweight='bold')
axes[0].set_xlabel('k'); axes[0].set_ylabel('Inertia')

axes[1].plot(K_range, sil_scores, 'o-', color='#F44336')
axes[1].set_title('Silhouette Score vs k', fontweight='bold')
axes[1].set_xlabel('k'); axes[1].set_ylabel('Silhouette Score')

plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'step7_elbow.png', bbox_inches='tight')
plt.show()
```

```python
# ── Run 3 clustering algorithms ───────────────────────────────
km3  = KMeans(n_clusters=3, random_state=42, n_init=10)
agg3 = AgglomerativeClustering(n_clusters=3)
db   = DBSCAN(eps=2.5, min_samples=10)

labels_km  = km3.fit_predict(X_clust)
labels_agg = agg3.fit_predict(X_clust)
labels_db  = db.fit_predict(X_clust)

n_db = len(set(labels_db)) - (1 if -1 in labels_db else 0)

print('Clustering Results:')
print(f'  K-Means        Silhouette: {silhouette_score(X_clust, labels_km):.3f} '
      f'| ARI vs age: {adjusted_rand_score(true_labels, labels_km):.3f}')
print(f'  Agglomerative  Silhouette: {silhouette_score(X_clust, labels_agg):.3f} '
      f'| ARI vs age: {adjusted_rand_score(true_labels, labels_agg):.3f}')
print(f'  DBSCAN         Clusters found: {n_db} '
      f'| Noise points: {(labels_db==-1).sum()}')
```
